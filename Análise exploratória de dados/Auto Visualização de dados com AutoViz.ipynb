# Auto Visualização de dados

Uma das principais etapas de um projeto de *machine learning* é a Análise exploratória e dentro dela há a visualização dos dados de forma gráfica e para isso utiliza-se diversas bibliotecas de visualização de dados como o *Plotly*, *Matplotlib*, *Seaborn*, *Bokeh* e até *Pandas* ou outras ferramentas de visualização mais específicas, como *Power BI*, *Tableau* e *Qlik Sense* dentre outras. 

Muitas vezes esse processo pode ser um pouco demorado, principalmente por exige, além do conhecimento básico em programação em *python*, o conhecimento das funções de cada uma dessas bibliotecas antes citadas. Como essa etapa gráfica é, geralmente, curta e serve para responder certas hipóteses levantadas no início do projeto e/ou extrair *insights*, seria uma alternativa automatizar esse processo.

Para isso há uma biblioteca de *auto visualização* chamada *autiviz* que usarei rapidamente nesse *notebook*.

Para instalar esse biblioteca usei o comando *pip install autoviz* e abaixo estarei importando e instanciando o módulo de *auto visualização*. Veja que nem a biblioteca *pandas* foi necessário importar.

## Dados

A base utilizada pode ser obtida [aqui](https://www.kaggle.com/adityadesai13/used-car-dataset-ford-and-mercedes) e trata de um conjunto de carros da marca *audi* (cerca de 10668 observações) e suas características (nove colunas).

## Importando e instanciando a *autoviz*

```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

Agora basta aplicar a biblioteca definindo a base de dados e a variável *target*.

Abaixo a biblioteca já realiza um pequeno relatório da base de dados com suas dimensões, quais os tipos de dados e quais são os preditores.

```python
AutoViz_Class().AutoViz('audi.csv',depVar='price')
```

Acima pode ser ver que forma gerados:

1) Relação gráfica por meio de uma *scatterplot* (gráfico de dispersão) das variáveis contínuas em relação a variável *target*. Podemos ver que não há uma relação entre *price* e *mpg*, mas entre *engineSize* e *price* há, pois a medida que a *engineSize* aumenta o preço também é maior;

2) *Pairwise-plot* de todas as variáveis;

3) Histograma (distribuição dos dados), *Boxplots*, *QQplot* e *Violinplot* de todas as variáveis contínuas. Aqui vemos que as variáveis contínuas não possuem uma distribuição normal, o que pode ser um problema na hora de passá-las para modelos de *machine learning* ou de *deep learning*, pois alguns desses modelos pressupõem normalidade nas variáveis. Isso mostra que essas colunas precisam de algum tipo de tratamento (normalização ou padronização). Outra coisa que pode-se notar é a presença de muitos *outliers* na variável *price* que pode ser ruim para o treinamento de um modelo, então essa coluna pode requerer um tratamento;

4) *Heatmap* (Mapa de calor) mostrando a correlação entre as variáveis. Podemos identificar quais as *features* mais adequadas para ser usadas em algum modelo;

5) *Pivot Table* mostrando a relação da variável *target* com mais de uma *feature* categóricas;

6) Gráficos de barras (*barplots*) para ver quais as relações médias entre várias outras variáveis e nossa variável alvo;

7) Por último, a base de dados completa e suas colunas.
